```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
strImg = "flor.jpg"
img = cv2.imread(strImg, cv2.IMREAD_GRAYSCALE)  #cuando sabermos que vamos a trabaja solo en grisse se agregar el parametro cv2.
```

```python
plt.imshow(img, cmap="gray") #agregar el campo apra dfinir grises = camp
plt.show()
```

```python
img.shape   #(734= alto, 1403=ancho,  3=referencia a 3 canales por que esta imagen es a color )
```

```python
# Dimensiones
alto = img.shape[0]
ancho = img. shape[1]
```

```python
#Arreglos donde va estar la salida
ecualizada = np.zeros((alto, ancho, 1)) #matriz de cero alto y ancho de imagen origina y solo escala de grisis
frecuencias = np.zeros((256)) #arreglo, posicion de 0 a 256 
```

```python
for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i, j]
        frecuencias[pixel] = frecuencias[pixel] + 1
```

```python
densidad = np.true_divide(frecuencias, (alto*ancho))  #el total de pixeles
```

```python
#hacemos la cdf
cdfFunc = np.cumsum(densidad, dtype=float)
```

```python
frecuencias  #posiciones de 0 a 255
```

```python
densidad # valor entre 0 y 1 ya para ver los valores de color
```

```python
cdfFunc  #su valor total es 1  # caso discrito
```

```python

for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i, j]
        newPixel = 255*cdfFunc[pixel] #Ecualizacion  #trasnformar el pixel
        ecualizada[i, j] = round(newPixel, 0) # se redodnde para quitar los decimales
```

```python
cv2.imwrite("ecualizada2.jpg", ecualizada)
```

```python
strImgEcualizada = "ecualizada2.jpg"
imgEcualizada = cv2.imread(strImgEcualizada, cv2.IMREAD_GRAYSCALE)
plt.imshow(imgEcualizada, cmap="gray")
plt.show()
```

```python
hist1 = cv2.calcHist([img], [0], None, [256], [0,256])
plt.plot(hist1, color="gray")
plt.xlabel("Intensidad de Pixeles")
plt.ylabel("Densidad")
plt.title("Imagen Original")
plt.show()
```

**Problema 1:**
Desarrolle una funcion la cual reciba dos parametros, una imagen y un entero llamado color, la funcion debe devolver una imagen la cual tenga activos los canales de color segun los siguientes puntos:

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("F1.jpg") #OpenCV lee en BGR
```

```python
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
#dimensiones
alto = img.shape[0]
ancho = img.shape[1]
```

```python
ancho
```

```python
#generar una estructura a donde escribir el resultado
lienzo = np.zeros((alto, ancho, 3)) #el resultado va a ser a color
```

```python
for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i, j]
        
        blue = pixel[2]
        green = pixel[1]
        red = pixel[0]
        
        lienzo[i, j] = [blue, 0, 0]
```

```python
cv2.imwrite("monocromo.jpg", lienzo)
monocromo = cv2.imread("monocromo.jpg")
monocromo = cv2.cvtColor(monocromo, cv2.COLOR_BGR2RGB)                       
```

```python
plt.imshow(monocromo)
plt.show()
```

```python
## Escala de Grises
```

```python
#dimensiones
alto = img.shape[0]
ancho = img.shape[1]
```

```python
media_ponderada = np.zeros((alto, ancho, 1)) #el resultado va a ser en escala de grises
```

```python
for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i, j]
        
        blue = pixel[2]
        green = pixel[1]
        red = pixel[0]
        
        media_ponderada[i, j] = int(0.299*blue + 0.587*green + 0.114*red)
```

```python
media_ponderada
```

```python
cv2.imwrite("gris_ponderado.jpg", media_ponderada)
gris = cv2.imread("gris_ponderado.jpg")
gris = cv2.cvtColor(gris, cv2.COLOR_BGR2RGB)
```

```python
gris.shape
```

```python
plt.imshow(gris)
plt.show()
```
